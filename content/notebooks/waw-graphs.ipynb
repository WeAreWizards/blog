```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import scipy, scipy.stats as st
plt.xkcd()
```

```python
x = np.linspace(0, 3.2, 100)
y = np.exp(x % 1)
yrisk = np.exp(x * 0.6)
fig, (ax0, ax1) = plt.subplots(1, 2, sharey=True, sharex=True)
fig.set_size_inches(12, 4)
ax0.set_ylabel(u'Risk of failure')
ax0.set_xlabel(u'Time (weeks of work)')
ax0.set_title('Small units of work')
ax1.set_title('Large unit of work')
plt.xticks([])
plt.yticks([])
ax0.annotate("Reset", (.6, 2.7))
l = ax0.plot(x, y)[0]
l.set_color('#4EB08F')
l = ax1.plot(x, yrisk)[0]
l.set_color('#4EB08F')
plt.savefig('../images/risk-vs-time.png')
```

```python
x = np.linspace(0, 10, 100)
kernel = st.gaussian_kde([0, 1,  2, 2, 3, 5, 10])
kernel2 = st.gaussian_kde([ 4, 9, 10.5, 10.5, 10.2, 11])
y = kernel.evaluate(x)
y2 = kernel2.evaluate(x)
fig, (ax0, ax1) = plt.subplots(1, 2, sharey=True, sharex=True)
fig.set_size_inches(12, 4)
ax0.set_ylabel(u'Cost of development')
ax0.set_xlabel(u'Time (weeks of work)')
ax0.set_title('Simple')
ax1.set_title('Easy')
plt.xticks([])
plt.yticks([])
ax1.annotate("ouch!", (8, .12))
ax0.annotate("bliss...", (8, .035))

plt.ylim((0, y.max() * 1.1))
l = ax0.plot(x, y)[0]
l.set_color('#4EB08F')
l = ax1.plot(x, y2)[0]
l.set_color('#4EB08F')
plt.savefig('../images/simple-vs-easy.png')
```

```python
x = np.linspace(0, 50, 100)
y = np.log(x + 10) - np.log(10.0)
y2 = ((x * 0.05)**2)
fig, (ax0, ax1) = plt.subplots(1, 2, sharey=True, sharex=True)
fig.set_size_inches(12, 4)
ax0.set_ylabel(u'Complexity')
ax0.set_xlabel(u'Number of components')
ax0.set_title('Good')
ax1.set_title('Bad')
plt.xticks([])
plt.yticks([])
plt.ylim((0, y.max() * 2.0))
ax0.plot(x, x*0.05)[0].set_color('#e0e0e0')
l = ax0.plot(x, y)[0]
l.set_color('#4EB08F')
ax1.plot(x, x*0.05)[0].set_color('#e0e0e0')
l = ax1.plot(x, y2)[0]
l.set_color('#4EB08F')
plt.savefig('../images/complexity-scaling.png')
```
